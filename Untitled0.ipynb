Sweet, so we know that this is a Series, and thus we can resample it.

```python
r_lecture_mean = ruby_lecture.T.apply(np.mean)
r_homework_mean = ruby_homework.T.apply(np.mean)
p_lecture_mean = python_lecture.T.apply(np.mean)
p_homework_mean = python_homework.T.apply(np.mean)
```

```python
r_lecture_mean.resample("W").plot(label="Ruby Lecture")
r_homework_mean.resample("W").plot(label="Ruby Homework")
p_lecture_mean.resample("W").plot(label="Python Lecture")
p_homework_mean.resample("W").plot(label="Python Homework")
plt.legend(loc=8)
plt.title("Weekly Average Difficulty")
plt.show()
```

So, this was pretty clunky getting here, but we have a(n unlabeled) graph of the difficulty through the average week. Thursday lectures tend to be harder than Monday lectures for the Python class. Neat.

```python
ruby_lecture["day_of_week"] = ruby_lecture.index.to_series().map(lambda d: d.weekday())
python_homework["day_of_week"] = python_homework.index.to_series().map(lambda d: d.weekday())
ruby_homework["day_of_week"] = ruby_homework.index.to_series().map(lambda d: d.weekday())
```

```python
ruby_lecture_weekdays = ruby_lecture.groupby("day_of_week")
python_homework_weekdays = python_homework.groupby("day_of_week")
ruby_homework_weekdays = ruby_homework.groupby("day_of_week")
```

```python
python_lecture_weekdays.apply(np.mean).T.mean().plot(label="Python Lecture")
ruby_lecture_weekdays.apply(np.mean).T.mean().plot(label="Ruby Lecture")
python_homework_weekdays.apply(np.mean).T.mean().plot(label="Python Homework")
ruby_homework_weekdays.apply(np.mean).T.mean().plot(label="Ruby Homework")
plt.legend(loc=9)
plt.show()
```
